```python
'''
Experiment: Stochastic Gradient Descent on Linear Regression

Goal
----
Verify that minibatch SGD can recover the correct parameter for a simple
linear relationship y = 3x.

This experiment simulates empirical risk minimization using a randomly
generated dataset and updates the parameter theta using minibatch
stochastic gradient descent.

Expected Result
---------------
theta should converge close to 3.
'''

import random
from sgd import SGD

# ------------------------------------------------
# Generate synthetic dataset
# ------------------------------------------------
# We simulate a simple linear function:
#
# y = 3x
#
# The learning algorithm does NOT know the true parameter (3).
# It must discover it using SGD.

data = []
for i in range(100):
    x = random.uniform(-10, 10)
    y = 3 * x
    data.append((x,y))

# ------------------------------------------------
# Generate synthetic dataset
# ------------------------------------------------
# We simulate a simple linear function:
#
# y = 3x
#
# The learning algorithm does NOT know the true parameter (3).
# It must discover it using SGD.

#start with a random parameter so the algorithm learns
theta = random.uniform(-5,5)

#-----------------------------------------------------
#Initialise model parameters
#-----------------------------------------------------

lr = 0.01 #learning rate
batch_size = 20
iterations = 200

#get the SGD optimizer
opt = SGD(lr)

#simulate algorithm optimization
for k in range(iterations):
    #sample m examples from the training data set
    batch = random.sample(data, batch_size)
    #declare gradient sum
    grad_sum = 0
    for x, y in batch:
        #predict value
        y_hat = theta * x
        #get the gradient loss
        grad_hat = 2 * x * (y_hat - y)
        #sum the gradient loss to the summ of gradients
        grad_sum += grad_hat
    #compute unbiased estimate of true gradient by taking average of the estimated gradients
    grad_avg = grad_sum / batch_size
    #compute parameter update using Stochastic Gradient Descent (SGD)
    theta = opt.update(theta, grad_avg)

    #track the value of the parameter over a # of iterations
    if (k % 20 == 0):
        print("iter: ", k, " theta=", theta)

print("final value of theta=", theta)

```
